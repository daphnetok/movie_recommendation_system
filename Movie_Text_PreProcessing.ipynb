##### Remove numbers and special characters 

```python
#remove numbers and special characters; only keep alphabets
pattern = '([^a-zA-Z ]+?)'

replace = ''

for ind in df.index:
    df['cleaned_tokens'][ind] = re.sub(pattern, replace, str(df['cleaned_tokens'][ind]))
```

##### Define stopwords

```python
stopwords = list(stopwords.words('english'))
```

##### Define lemmatizer

```python
lemmatizer = WordNetLemmatizer()
```

##### Word tokenization, Remove stopwords, Lemmatization.

```python
#nltk.download('omw-1.4')
```

```python
for ind in df.index:
    #tokenization
    tokens = mwe_tokenizer.tokenize(word_tokenize(str(df["cleaned_tokens"][ind])))
    #create list to temporarily store cleaned tokens later to append into new column 'cleaned_tokens'
    myList = []
    for i in tokens:
        #if token is not a stopword, lemmatize then append into list
        if i not in stopwords:      
            myList.append(lemmatizer.lemmatize(i, pos='v'))

    #put tokens list back into df by index
    df['cleaned_tokens'][ind]=myList
```

#### Lemmatizating the cleaned tokens by POS tagging each token. 

Because the default NLTK lemmatization only accepts 1 tag as arguement, we have to define a function in order to lemmatize all the different types of tokens (Nouns, Verbs, Adjectives, Adverbs)

```python
def get_wordnet_pos(tag):
    
    #if token is a noun
    if tag.startswith('N'):
        return wordnet.NOUN

    #if token is a verb
    elif tag.startswith('V'):
        return wordnet.VERB
        
    #if token is an adjective
    elif tag.startswith('J'):
        return wordnet.ADJ
    
    #if token is an adverb
    elif tag.startswith('R'):
        return wordnet.ADV

    #if the token is already in its root form
    else:
        return wordnet.NOUN # default return value for all other POS tags
```

##### Lematize tokens by finding their pos tag (NN, VB, JJ or RB) for every row in df, make new array to store lemmatized words

```python
for ind in df.index:
    tokens_lem = []
    
#tag each token to their pos tag
    pos = nltk.pos_tag(df['cleaned_tokens'][ind])
#     print(pos)
    #apply the function to each token to get their pos tag & compile result into a list
    wordnet_tagged = list(map(lambda x: (x[0], get_wordnet_pos(x[1])), pos))
    #for each token in column, lemmatize according to their pos tag
    for word, tag in wordnet_tagged:
        
        tokens_lem.append(lemmatizer.lemmatize(word, tag))
        
        #append the list with lemmatized tokens into the df
        df['cleaned_tokens'][ind]=tokens_lem

```

#### Spell checker to check if cleaned token is spelled correctly and correct them if it is wrong

```python
from spellchecker import SpellChecker

def spell_checker(df, cleaned_tokens):
    spell = SpellChecker(language='en')
    for ind in df.index:
        tokens = df[cleaned_tokens][ind]
        corrected_tokens = []
        for token in tokens:
            corrected_token = spell.correction(token)
            corrected_tokens.append(corrected_token)
        df[cleaned_tokens][ind]= corrected_tokens
    return df
```
